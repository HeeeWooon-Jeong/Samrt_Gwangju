# 목표 설정

- 논리연산자 and와 xor 값을 예측해보자

```python
import pandas as pd
```

```python

# and 게이트 만들기 - DataFrame을 이용해서 논리게이트를 만들어보자
and_data = {
         'A' : [1,0,0,1,1,0,0,1],
         'b' : [0,0,1,1,0,1,1,1],
    'A and B': [0,0,0,1,0,0,0,1]
}

df_and = pd.DataFrame(and_data)
df_and
```

```python
# 머신러닝을 진행하기 위한 데이터 분할
# 문제와 정답 데이터를 분리

X = df_and.iloc[ : , : 2 ] # 인덱서 사용
y = df_and['A and B'] # 열 인덱싱


y
```

```python
print(X.shape)
print(y.shape)
```

```python
# 훈련 데이터셋 / 정답 데이터셋으로 분리
X_train = X.iloc[ 0 : 6, : ]
X_test = X.iloc[ 6 : , : ]
y_train = y[0 : 6]
y_test = y[6:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# 08/31

```python
# 머신러닝 모델을 사용해서 학습과 평가 / 예측을 진행해보자
# sklearn - 머신러닝에 관련된 모델 / 유용한 기능 / 데이터가 모여잇는 라이브러리
```

```python
# 모델 import - 모델 = 이미 구현된 알고리즘
# 

from sklearn.neighbors import KNeighborsClassifier
```

```python
# 모델 객체 생성 - 모델을 사용하기 쉽도록 튜닝하고 변수에 저장
knn_and = KNeighborsClassifier(n_neighbors=1)
```

```python
# 모델 학습 - fit(훈련용 문제데이터, 훈련용 정답데이터)
knn_and.fit(X_train, y_train)
```

```python
# 경고창 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 모델 데이터 예측 - predict()
pre = knn_and.predict(X_test) #문제를 풀었던 답안지
pre
```

```python
# 모델 예측 데이터 평가
from sklearn.metrics import accuracy_score # 정확도
```

```python
accuracy_score(y_test, pre)
```

```python

# xor 게이트 만들기 - DataFrame을 이용해서 논리게이트를 만들어보자
xor_data = {
         'A' : [1,0,0,1,1,0,0,1],
         'b' : [0,0,1,1,0,1,1,1],
    'A xor B': [1,0,1,0,1,1,1,0]
}

df_xor = pd.DataFrame(xor_data)
df_xor
```

```python
X = df_xor.iloc[ : , : 2 ] # 인덱서 사용
y = df_xor['A xor B'] # 열 인덱싱


y
```

```python
# 훈련 데이터셋 / 정답 데이터셋으로 분리
X_train = X.iloc[ 0 : 6, : ]
X_test = X.iloc[ 6 : , : ]
y_train = y[0 : 6]
y_test = y[6:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델 객체 생성
knn_xor = KNeighborsClassifier(n_neighbors=1)
```

```python
# 모델 학습
knn_xor.fit(X_train,y_train)
```

```python
# 모델 데이터 예측
pre = knn_xor.predict(X_test)
```

```python
# 모델 데이터 예측 평가
# score(평가용 문제 데이터, 평가용 정답 데이터) : 간단하게 모델의 성능을 평가하는 함수
knn_xor.score(X_test, y_test)
```
