```python
#!pip install missingno
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import missingno as msno
from sklearn.tree import DecisionTreeClassifier
```

```python
train = pd.read_csv("train_09_15_1.csv")
X_test = pd.read_csv("test_09_15_1.csv")
y_train= train['Reached.on.Time_Y.N']
X_train = train.drop('Reached.on.Time_Y.N', axis=1)
 
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# XGBoost와 LightGBM의 모델과 하이퍼파라미터 범위 정의
xgb_param = {
    'learning_rate': [0.006],
    'max_depth': [ 5],
    'n_estimators': [45],
    'min_child_weight': [10]
    
}
lgbm_param = {
    'max_depth': [20],
    'n_estimators': [7]
}
# XGBoost 모델 정의
xgb_model = XGBClassifier(random_state=10)

# GridSearchCV를 사용하여 XGBoost 모델 튜닝
xgb_grid = GridSearchCV(xgb_model, xgb_param, cv=65, scoring='accuracy', n_jobs=-1, verbose=3)
xgb_grid.fit(X_train, y_train)

# 최적의 하이퍼파라미터와 정확도 출력
print("XGBoost 최적 하이퍼파라미터:", xgb_grid.best_params_)
print("XGBoost 최고 정확도:", xgb_grid.best_score_)

# LightGBM 모델 정의
#lgbm_model = LGBMClassifier(random_state=10)

# GridSearchCV를 사용하여 LightGBM 모델 튜닝
#lgbm_grid = GridSearchCV(lgbm_model, lgbm_param, cv=65, scoring='accuracy', n_jobs=-1)
#lgbm_grid.fit(X_train, y_train)

# 최적의 하이퍼파라미터와 정확도 출력
#print("LightGBM 최적 하이퍼파라미터:", lgbm_grid.best_params_)
#print("LightGBM 최고 정확도:", lgbm_grid.best_score_)
# 최적의 모델과 하이퍼파라미터를 사용하여 테스트 데이터에 대한 정확도 평가
xgb_best_model = xgb_grid.best_estimator_
#lgbm_best_model = lgbm_grid.best_estimator_

from sklearn.model_selection import cross_val_score
cross_val_score(xgb_best_model, X_train, y_train, cv=12).mean()

#cross_val_score(lgbm_best_model, X_train, y_train, cv=12).mean()

pre = xgb_best_model.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score

# 모델 생성
gbm_model = GradientBoostingClassifier()

# 하이퍼파라미터 그리드 설정
param_grid = {
    'n_estimators': [28],
    'max_depth': [4],
    'min_samples_split': [2]
}

# GridSearchCV를 사용하여 최적 하이퍼파라미터 탐색
grid = GridSearchCV(gbm_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=3)
grid.fit(X_train, y_train)

# 최적 하이퍼파라미터 출력
print("최적 하이퍼파라미터:", grid.best_params_)

# 최적 모델로 교차 검증 수행
cross_val_scores = cross_val_score(grid.best_estimator_, X_train, y_train, cv=5)
print("교차 검증 평균 정확도:", cross_val_scores.mean())

# 모델 예측
pre3 = grid.predict(X_test)
```
