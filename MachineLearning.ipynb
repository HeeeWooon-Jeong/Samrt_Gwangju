통계청 공공기관포털 aihub 

모집단(population) 과 표본(Sample)

    모집단 : 관심의 대상이 되는 모든 개체의 관측값이나 측정값의 집합
    ex:7강의실 남자의 키
    표본 : 모집단에서 추출한 일부 관측값이나 측정값의 집합
    ex:2분단 남자의 키
    
전수조사 : 모집단 전체를 조사하는 것
    신뢰도 높음

표본조사 : 표본만 가지고 조사하는 것  / 표본을 잘 잡아야함
    시간 비용 단축



    개체 : 연구자 또는 관찰자가 관심을 갖는 대상  
        - 
    요인 : 개체에 관한 특성 중 연구자가 특별히 관심을 갖는 특성  
        - 스펙

    변수 : 요인을 구성하고 있는 요소
        - 학교성적, 자격증, 개발능력, 수상이력 ,프로젝트 경험
        


독립변수(Independent variable)  
        다른 변수에 영향을 받지 않고, 종속변수에 영향을 주는 변수
    
    
종속변수(Dependent variable)  
        종속 또는 의존적인 변수, 독립변수에 영향을 받아서 변화하는 변수


## 종속변수 (출력값)        y = 3x       독립변수(입력값)
