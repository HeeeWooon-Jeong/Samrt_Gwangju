```python
# 라이브러리 import
import pandas as pd
```

```python
# 데이터 프레임 만들기(딕셔너리) - 인공지능
df_1 = pd.DataFrame(data = {'진로희망' : ['빅데이터분석가','자영업','웹개발자','대학원','합계'],
                           '도수' : [6, 7 , 4, 3, 20]})
df_1

```

### 도수

- 빈도 수 (전체 데이터 중에서 해당하는 데이터의 갯수)


```python
# 데이터 프레임 만들기(딕셔너리) - 빅데이터
df_2 = pd.DataFrame(data = {'진로희망' : ['빅데이터분석가','자영업','웹개발자','대학원','합계'],
                           '도수' : [12, 14 , 10, 14, 50]})
df_2
```

```python
# 빅데이터 반에 있는 도수를 가져오자
df_col = df_2['도수']
df_col
```

```python
df_1['빅데이터반'] = df_col
df_1
```

```python
df_1.set_index('진로희망')
```

```python
df_1.set_index('진로희망',inplace=True)
df_1
```
