#### 조건에 맞는 데이터 접근하기
- 불리언 인덱싱

```python
# random.choice : 랜덤으로 하나의 요소를 취득
# random.sample : 랜덤으로 여러요소를 취득, 중복 없음
# random.choices : 랜덤으로 여러 요소를 취득, 중복 있음
```

```python
# 불리언 배열 생성
sample_arr = [True,False]

np.random.seed(1)
bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_data = np.arange(0,10) # 0 ~ 9 까지 1씩 증가하는 숫자 생성
display(tmp_data)
```

```python
tmp_data[bool_arr]
```

```python
# 논리연산자를 활용하여 80점 이상인 성적 확인하기
score[score >= 80]
```

```python
# 70 점 미만인 데이터는 몇개 일까?
e = score[score < 70]
e
```

```python
e.size
```

```python
# 학생들의 번호 생성
std_num = np.arange(1,21)
std_num
# 성적이 60점 이상인 학생들의 번호 출력하기
std_num[score>=60]
```

#### 데이터 불러오기

```python
data = np.loadtxt("./data/height_weight.txt", delimiter = ",")
data
```

```python
# 데이터 정보 확인하기(크기,데이터타임,차원)

# array 배열 크기
display(data.shape)
#데이터 타입
display(data.dtype)
#차원 확인
display(data.ndim)
```

```python
# 1.키 변수에 키 데이터담기

height = data[0]
height
```

```python
# 2. 몸무게 변수에 몸무게 데이터 담기
weight = data[1]
weight
```

```python
# 3. 키 데이터에의 단위 변경하기 cm --> m(배열의 요소가 연산)
height1 = height / 100
height1
```

```python
# 4. 배열간의 연산을 통새ㅓ bmi 지수 구하기(공식적용) : 몸무게 / 키(m) * 키(m)
total = weight // (height1 * height1)
total
```

```python
# 5. 과체중과 비만인 데이터들에 접근하기(불리언 인덱싱 활용)
display(total[total >= 23])

# 10명중 몇명일까?
display(total[total >=23].size)
```

```python
# 과체중인 데이터들에 접근하기
# & : 비트논리 연산자 and
# | : 비트논리 연산자 or
total[(total >= 23) & (total < 25)]
```

#### 예제 - 감귤나무 경계 판단


```python
# sqrt() 함수를 이용
tree = (3, 0)
tree_x, tree_y = tree[0],tree[1]
distance = np.sqrt((tree_x-0)**2 + (tree_y - 0)**2)
print(distance)

if distance == 2.0 :
    print("경계-on")
elif distance < 2.0 :
    print("내부-in")
else :
    print("외부-out")
```

### 영화 평점데이터 분석 - 각 사용자별 평점 평균 구하기

#### 데이터 불러오기
- data 변수에 담기
- np.loadtxt('경로',구분자,데이터타입)

```python
data = np.loadtxt('./data/ratings.dat',delimiter='::',dtype = np.int64) # int64 = 2^64
data
```

#### 데이터 정보 확인하기

```python
# 차원
print(data.ndim)
# 크기
print(data.shape)
# 요소 개수
print(data.size)
```

#### 전체 평점 평균 구하기

```python
rating_all = data[:,2].mean()
rating_all
```

```python
#사용자 아이디 접근
user_ids = data[:,0]
user_ids
```

```python
# 사용자 아이디 중복없이 확인
user_id = np.unique(user_ids)
user_id
```

```python
# 사용자 아이디가 1인 데이터의 평점 평균 구해보기
user_id1 = data[data[:,0] == 1]
user_id1
```

```python
user_id1[:,:2].mean()
```

```python
# 모든 사용자의 평점에서 접근하여 평점 평균 구하고 저장하기
result = []
for i in user_id :
    rating_mean = data[data[:,0] == i][:,2].mean()
    result.append([i,rating_mean])
```

```python
result
```

```python
# csv 파일로 저장
np.savetxt('./data/user_rating_mean.csv',result,delimiter=',',fmt='%.3f')
```

```python
# 저장한 파일 다시 불러오기
user_rating_mean = np.loadtxt('./data/user_rating_mean.csv',delimiter=',')
user_rating_mean
```

```python
# 평점평균이 4점 이상이 사용자는 몇명인지 알아보기
user_rating_mean[user_rating_mean[:,1] >= 4.0][:,0].size
```
