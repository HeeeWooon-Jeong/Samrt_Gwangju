### 문자열 함수

##### 복습
- find(), index() : '찾을 문자열' 위치 반환<br>
=차이점<br>
    - find(): 문자를 찾을 수 없다면 -1 반환<br>
            error를 발생시키지 않으므로 더 많이 사용됨<br>
    - index(): 문자를 찾을 수 없다면 에러를 발생<br>
    - len() : 문자열 개수(길이)반환<br>
            공백!?,. 모두 포함<br>
    - replace('찾을 문자열' , '변경할 문자열')<br>
    : 찾을 문자열을 찾아서 변경할 문자열로 변강하는 함수<br>
    - count('찾을 문자열') : '찾을 문자열' 이 몇 개 있는지 개수를 반환<br>
    - split():사용자가 입력한 특정 기준으로 문자열을 나눠주는 함수<br>
    - strip(): 공백 제거 <br>
    - lstrip():왼쪽 공백제거<br>
    
    
- %기호 포매팅<br>
- format 함수를 사용한 포매팅<br>
- f문자열 포매팅<br>

<연산자><br>
- 산술연산자 : //순수한 몫<br>
- +, * 연산자를 활용하여 문자열이 반복되는 연산<br>
- 파이썬에서 형변환 <br>
  int(대상, 변수), str(대상, 변수)<br>

```python
cnt_mask = int(input('마스크 개수 입력 >> '))



rest = cnt_mask%8%5

if cnt_mask >= 8 :
    p8 = cnt_mask//8
    p5 = cnt_mask%8//5
    if rest > 0 :
        p5 += 1
elif cnt_mask < 8 :
    p8 = 0
    p5 = cnt_mask%8//5
    if rest > 0 :
        p5 += 1
    
print('8개들이 포장지 개수 : ', p8)
print('5개들이 포장지 개수 : ', p5)

```

```python
num1, num2 = map(int, input().split(' '))

if num1 > num2 :
    print('첫 번째 정수가 더 큽니다.')
elif num1 < num2 :
    print('두 번째 정수가 더 큽니다.')
else :
    print('같습니다.')
```

```python
score = int(input("점수 입력 >> "))


if score >= 90:
    print(score,'점은 A학점 입니다.')
elif score >= 80 :
    print(score,'점은 B학점 입니다.')
elif score >= 70 :
    print(score,'점은 C학점 입니다.')
elif score >= 60 : 
    print(score,'점은 D학점 입니다.')
else : 
    print(score,'점은 F학점 입니다.')
```

```python
# 1. 메뉴 출력
# 2. 돈 입력
money = int(input('금액넣기 >> '))
# 3. 선택 메뉴 입력
menu = int(input('메뉴 선택(번호입력) >> '))
# 4. 결과 확인
#  - 잔액부족 출력
#  - 거스름돈 환전(1000원, 500원, 100원)
change = 0 # 잔액 변수

# 1000원, 500원, 100원 잔액에 대한 반환금액 변수
ten, five, one = 0, 0, 0

if menu == 1 :
    change = money - 600
elif menu == 2 :
    change = money - 800
elif menu == 3 :
    change = money - 1000
    
if change > 0 :
    ten = change // 1000 # 1000원 몫
    five = change % 1000 // 500  #500원 몫
    one = change % 1000 % 500 // 100 # 100원 몫
    print(f'1000원 : {ten}장, 500원 : {five}개, 100원 : {one}개')
else : 
    print('금액이 부족합니다.')
```
