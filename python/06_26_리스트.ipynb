### 리스트(list)
- 파이썬 자료구조 중 하나 
- 순서가 있고, 수정이 가능한 자료구조
- 대괄호 ([ ])로 작성되고. 리스트 내부의 값은 콤마(.)로 구분
- 하나의 리스트에는 다양한 자료형을 포함
-추가, 수정, 삭제 가능

#### 리스트 선언

```python
# 리스트명 = [요소1, 요소2, ...]
a = [] # 빈 리스트 선언
b = [1, 2, 3] # 숫자요소를 담은 리스트 
c = ['봉민혁', '김현중', '김다영'] #문자열 요소를 담은 리스트
d = [1, 2, '정희운', '인여진', '박필수'] #숫자, 문자열 요소를 담은 리스트
e = [1, 2, '김유성', ['신지훈', '구동현']] #리스트 내부에 리스트를 포함한 리스트

# [] : 단일 리스트
# [[]] : 이중 리스트
```

#### 리스트 슬라이싱

```python
# 리스트명 [시작 인덱스 : 끝인덱스]
# 인덱스범위 : 시작인덱스 <= 인덱스번호 < 끝인덱스
list3 = ['이우석', '김동영', '장재욱', '강진구', '신지훈']
# '이우석' ~ '장재욱' 데이터 접근 출력
print(list3[0:3])
print(list3[:3])
```

```python
# 김동영 ~ 신지훈
list3[1:]
```

```python
# 인덱스 번호 0부터 끝까지 2개씩 띄어서 가져오기
# [시작인덱스 : 끝인덱스 : 증감값]
# 인덱스 번호 1부터 끝까지 3개씩 띄어서 가져오기
list3[1::3]
```

```python
# 이중리스트를 만들고 데이터 접근해보기
# 이름, 나이, 혈액형 정보기반의 이중 리스트
list4 = [['김준우', 29, 'B'],
         ['정희운', 30, 'B'],
         ['봉민혁', 24, 'A'],
         ['김유성', 29, 'B'],
         ['강진구', 26,'O']]

# 봉민혁의 나이, 혈액형 접근해서 출력
# 민혁이의 나이는 oo 이고, 혈액형은 o형 입니다
name = input()
for i in range(len(list4)) :
    if name == list4[i][0] :
        age = list4[i][1]
        blood = list4[i][2]


print(f'{name}의 나이는 {age} 이고, 혈액형은 {blood}형 입니다.')
```

```python
# 김유성, 강진구 데이터에 접근해서 이름, 나이 정보 출력하기
# oo의 나이는 o살 입니다.

list4[3:]
print(f'{list4[3][0]}의 나이는 {list4[3][1]}살 입니다.')
```

### tuple은 값 추가,수정,삭제 불가
 - 파이썬의 자료구조 형태 중 하나
 - 순서가 있는 집합(인덱싱, 슬라이싱 가능)
 - 소괄호 ()로 작성되어지고, 튜플 내부의 값은 콤마(.) 로 구분
 - 하나의 튜플에는 다양한 자료형을 포함
 - 데이터의 추가, 수정, 삭제가 불가능
 - 프로그램이 실행되는 동안 요소값이 항상 변하지 않아야 할때
 


```python
list = ['딸기', '바나나', '복숭아', '체리', '사과']

serch =input('검색할 문자를 입력하세요 >> ')
fruit = serch
if fruit in list :
    index = list.index(fruit)
    print(f'{fruit}는 리스트 {index}번째 인덱스에 들어있습니다.')
else :
    print(f'{fruit}는 리스트에 들어있지 않습니다.')

```

#### 패킹 과 언패킹



```python
def factorial(n):
    if n== 1:  # n이 1일 때
        return 1 # 1을 반환하고 재귀호출 끝냄
    return n * factorial(n-1) # n과 factorial 함수에 n - 1 을 넣어서 반환된 값을 곱함
 
print(factorial(5))
```
