### while 문
- 반복횟수가 명확하지 않을 때

```python
# num 이라는 변수를 선언하고
# 파이썬 최고!!! 13번 출력
num = 1

while num <= 13:
    print(f'{num} : 파이썬 최고')
    num += 1
```

```python
# break

while True :
    print('출력')
    break
```

```python
# continue

num = 0

while num < 10 :
    num += 1 #num = num + 1
    if num % 2 == 0:
        continue
    print(num)
```

```python
num3 = num1 + num2

while True :
    
    if num1 == 0 and num2 == 0 :
        
        print('프로그램이 종료되었습니다')
        break
    else :
        num1 = int(input('첫번째 정수 입력 >> '))
        num2 = int(input('두번째 정수 입력 >> '))
        print(f'두 정수의 합 :{num3}')
```

```python
num1 = int(input('현재 몸무게 : '))
num2 = int(input('목표 몸무게 : '))
num3 = 1
total = 0
num1 - total
while True :
    num4 = int(input(f'{num3}주차 감량 몸무게:'))
    num3 += 1
    total += num4
    if num1 - total < num2 :
        num1 -= total
        print(f'{num1}kg 달성 축하합니다')
        break

```

### for 문 
- 반복횟수가 정해저 있을때

```python
lunch = ['큐브 식판밥','김치찌개','한솥','돈까스']
lunch
```

```python
for i in lunch :
    print(i)
```

```python
hi = '안녕하세요'
for i in hi :
    print(i,end = '' ) # print()속성 : 기본값은 개행해서 출력
                       # shift + tab 으로 함수 속성 확인
                       # 출력문의 끝 속성 : end = "" or ''
```

```python
score_list = [90, 45, 70, 60, 55]
num = 0
for i in score_list :
    num += 1
    if i >= 60 :
        print(f'{num}번째 학생은 합격입니다')
    else :
        print(f'{num}번째 학생은 불합격입니다')
```

```python
score_list = [90, 45, 70, 60, 55]

for i in score_list :
    if i >= 60 :
        print(f'{score_list.index(i)+1}번 학생은 합격 입니다.')
    else :
        print(f'{score_list.index(i)+1}번 학생은 불합격 입니다.')
```

```python
for i in range(10, 0, -1):
    print(i)
```

#### for : range()
- 필요한 만큼의 숫자를 만들어 내는 함수
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
lunch
```

```python
for i in range(1, len(lunch)) :
    print(i, lunch[i])
```

```python
score_list = [90, 45, 70, 60, 55]

for i in range(len(score_list)) :
    if score_list[i] >= 60 :
        print(f'{i+1}번 학생은 합격 입니다.')
    else :
        print(f'{i+1}번 학생은 불합격 입니다.')
```

```python
for i in range(1, 25, 3) :
    print(i, end=' ')
    
```

### MIN, MAX 함수
- 선언한 자료구조(리스트, 튜플)에서 가장 큰 값, 가장 작은 값을 찾아주는 함수

```python
list2 = [4,5,2,56,7,8,96,48,5,12,88]
#list2 에서 가장 작은 수 구하기
min(list2)

#list2 에서 가장 큰 수 구하기
max(list2)
```

```python
# 가장 큰 수 구하기

max_num = list2[0] 

for i in list2 :
    if max_num < i :
        max_num = i
print(max_num)
```

```python
result = 0

for i in range(0, 100, 3) :
    result += i
print(result)
```

```python
# 1 ~ 100 사이의 숫자중 3의 배수의 합을 구하기

add = 0
for i in range(1,101) :
    if i%3==0 :
        add += i
print(add)
```

```python
for i in range(1, 10) :
    print('2 *',i,'=',2*i)
```

#### 반복문   
##### While : 반복횟수가 명확하지 않을 때  
break : 반복문을 강제 종료  
continue : 어떤 특정 조건을 제외시킬 때 사용  
           continue를 만나면 continue아래 코드 무시되고 다음 반복문 실행  

###### for : 반복횟수가 명확할때
range(시작값, 종료값, 증감값) : 필요한 만큼의 숫자를 만들어내는 함수  
시작할 숫자가 0이라면 생략 가능 - 기본값  
증감값이 1씩 증가하는 경우 생략 가능 - 기본값  
ex) range(11) -> 0~10까지 1씩 증가하는 숫자 생성
