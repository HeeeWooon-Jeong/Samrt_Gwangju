```python
import pickle
```

```python
# 멜론차트의 데이터를 불러와서 데이터 확인하기
# with문 사용해서 데이터 불러오기

with open('music_100.pkl','rb') as f : # 'rb' = read binary
    music = pickle.load(f)
```

```python
music[:5]
```

```python

s_song= input('제목 중 찾고 싶은 키워드를 입력하세요 >> ')
s_cnt = 0 # 찾은 노래 곡수 담을 변수
song_list = [] #찾은 노래의 순위정보와 노래제목을 담을 빈 리스트


# 인덱스 0 : 차트 순위, 1 : 노래 제목 , 2 : 가수이름
for i in music :
    
    if s_song in i[1] :
        s_cnt += 1
        song_list.append([i[0], i[1]])      
print(f'\'{s_song}\'으로 검색한 노래 수 : {s_cnt}곡')

print(f'순위와 노래 제목 : {song_list}')
```

```python
# 특정 가수의 노래 만들기

# 1. 검색할 가수명 입력받기

# 2. for 반복문을 이용하여 전체 데이터를 가져오기

# 3. for 반복변수가 꺼내온 정보 확인
#    내가 입력받은 가수명이 꺼내온 정보의 어디에 있는지 확인

# 4. 데이터가 있다면 노래리스트에 담아주기

# 5. 노래 리스트 출력하기

```

```python
list1 = []
serch = input('검색할 가수명 >> ')

for i in music :
    if serch in i[2] :
        list1.append([i[0],i[1],i[2]])
        
print(list1)
```

```python
list1 = []
serch = input('검색할 가수명 >> ')
for i in music :
    if serch in i[2] :
        i[2]= '정희운'
        list1.append([i[0],i[1],i[2]])
        
list1
```

```python
my_song = []

cnt = 0

for i in music :
    if i[2] == '정희운' :
        i[2] = '정희운'
        my_song.append()
        cnt +=1
print(f'melon 차트속 나의 노래 : 총 {cnt}곡')
my_song
```
