```python
# 전체 리뷰길이 평균값
cnt= 0
for i in range(len(movie)) :
    cnt += len(movie[i][0])
    
print(cnt/len(movie))

```

```python
# 긍정 / 부정 리뷰의 개수를 구해보자

cnt_t = 0
cnt_f = 0
for i in movie :
    
    if i[1] == 1 :
        cnt_t += 1
        
    else :
        cnt_f += 1
        
# 긍정        
print(cnt_t) 

# 부정
print(cnt_f)
    
  
```

```python
# 리뷰길이의 최대 / 최소값을 확인하고 그 내용을 출력하기

# 1. min max함수를 사용하여 최대 / 최소값 확인하기\
length = []
for i in movie :
    length.append(len(i[0]))
print(max(length))

length = []
for i in movie :
    length.append(len(i[0]))
print(min(length))




# 3. 최소 길이의 리뷰 뽑아서 출력 - 개수도 함께 확인


```

```python
max_num = max(length)
min_num = min(length)

max_review = [] # 최대길이 리뷰를 담을 빈 리스트
max_cnt = 0
for i in movie :
# 2. 최대 길이의 리뷰 뽑아서 출력하기 - 개수도 함께 확인
    if max_num == len(i[0]) :
        max_review.append(i[0])
        max_cnt += 1
# 3. 최소 길이의 리뷰 뽑아서 출력하기 - 개수도 함께 확인
    if min_num == len(i[0]) :
        min_review.append(i[0])
        min_cnt += 1

print(f'최대 길이 리뷰 수는 {max_cnt}이고, 내용은 = {max_review}')
print()
print(f'최소 길이 리뷰 수는 {min_cnt}이고, 내용은 = {min_review}')
```

```python
n = 6  # 패턴의 크기
for i in range(-n+1, n):  # 패턴의 행 인덱스 범위 (-n+1부터 n-1까지)
    spaces = abs(i)  # 공백의 개수는 행 인덱스의 절댓값과 동일
    stars = 2*(n - abs(i)) - 1  # 별의 개수는 행 인덱스와 관련
    line = ' ' * spaces + '*' * stars  # 공백과 별을 결합하여 한 줄 생성
    print(line)
```
