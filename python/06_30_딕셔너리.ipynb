

### 딕셔너리란 ?
- Key와 Value를 한쌍으로 갖는 자료형  
- 불변하는 Key와 가변/불변적인 Value로 맵핑된 순서가 없는 집합
- 인덱싱, 슬라이싱 불가  

### 딕셔너리 선언

```python
# 딕셔너리명 = {key : value, key : Value, ....}
a = {} # 빈 딕셔너리
b = {'name' : 'hw'} #이름정보(문자열) 담은 딕셔너리
c = {1 : 'ey', 2 : 23} #하나의 딕셔너리에는 다양한 자료형을 담을 수 있음  
d = {'name' : ['우석', '동영', '유성']} # value 값으로 리스트로 묶어서 담을 수 있음   


# 명확하게 뭔가를 정의하고 싶을 때 사용  
```

```python
print(a)
print(b)
print(c)
print(d)
```

#### 딕셔너리 데이터 접근 방법 비교
- 딕셔너리명[키]vs딕셔너리명.get(키)

```python
# 딕셔너리명[키]로 접근할 때
dic['blood']
```

```python
a = dic.get('blood')
print(a)
```

```python
# 예외 처리 : try ~ except
try :
    result = dic['blood']
    result
except KeyError:
    print('없는 키 값 입니다.')
    
```

#### 딕셔너리 key, value 모두 가져오기
- items()

```python
dic.items()
```

```python
list(dic.items())
```

```python
# 패킹된 형태로 출력
# 튜플 형태로 출력
for i in dic.items() :
    print(i)
```

```python
# 언패킹된 형태로 출력
for key, value in dic.items() :
    print(key, value)
```

```python
score_dict = {'이름' : ['성춘향', '이몽룡', '김방자', '최향단', '신지훈'],
             'Python': [90, 100, 85, 90, 80],
             'java' : [85, 100, 80, 95, 85],
             'html/css' : [75, 70, 90, 80, 90]}
list(score_dict.values())[1]
sum(list(score_dict.values())[1])/len(list(score_dict.values())[1])
```

```python
sum(list(score_dict.values())[1])/len(list(score_dict.values())[1])
```

```python
print('Python 평균 점수 : ', end='')
print(sum(list(score_dict.values())[1])/len(list(score_dict.values())[1]))
print('Java 평균 점수 : ', end='')
print(sum(list(score_dict.values())[2])/len(list(score_dict.values())[2]))
print('Html/CSS 평균 점수 : ', end='')
print(sum(list(score_dict.values())[3])/len(list(score_dict.values())[3]))
```

```python
score_dict['Python']
```

```python
sum(score_dict.value())
```

```python
score_dict.values()
```

```python
# 이름은 빼고 나머지 과목만 가져오기 위해 리스트로 형변환 --> 인덱싱, 슬라이싱을 위해

for i in list(score_dict.keys())[1:] :
    avg = sum(score_dict[i])/len(score_dict[i])
    print(f'{i} 평균 : {avg}')
```
