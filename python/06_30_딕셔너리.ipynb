```python
score_dict = {'' : ['', '', '', '', ''],
             'Python': [90, 100, 85, 90, 80],
             'java' : [85, 100, 80, 95, 85],
             'html/css' : [75, 70, 90, 80, 90]}
list(score_dict.values())[1]
sum(list(score_dict.values())[1])/len(list(score_dict.values())[1])
```

```python
sum(list(score_dict.values())[1])/len(list(score_dict.values())[1])
```

```python
print('Python   : ', end='')
print(sum(list(score_dict.values())[1])/len(list(score_dict.values())[1]))
print('Java   : ', end='')
print(sum(list(score_dict.values())[2])/len(list(score_dict.values())[2]))
print('Html/CSS   : ', end='')
print(sum(list(score_dict.values())[3])/len(list(score_dict.values())[3]))
```

```python
#         --> ,  

for i in list(score_dict.keys())[1:] :
    avg = sum(score_dict[i])/len(score_dict[i])
    print(f'{i}  : {avg}')
```
