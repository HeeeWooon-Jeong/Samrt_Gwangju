보조기억장치
- 주기억 장치에 비해 속도는 느리지만 전원이 차단되어도 내용 유지, 용량이 큼
- 중앙처리장치와 직접 자료 교환이 불가능
- 접근 시간이 오래 걸림
- 일반적으로 주기억장치에 데이터를 저장할 때는 DMA 방식을 사용
- CPU 가 직접 접근할 수 없다.


채널은 해독, 지시, 실행 기능이 이씅며 데이터 저장하는건 기억장치의 몫

ALU의 위치와 기능이 바르게 나열된것  
ALU(Arithmetic Logic Unit) : CPU 산술 논리 연산 장치




    캐시 메모리는 CPU 처리속도와 주기억장치의 접근 속도 차이를 줄이기 위해 사용하는 고속 Buffer Memory  
    
    캐시는 주기억장치와 CPU사이에 위치하며, 자주 사용하는 프로그램과 데이터를 기억
    
    캐시메모리는 메모리 계층 구조에서 가장 빠른 소자이며, 처리속도가 거의 CPU속도와 비슷할 정도의 속도를 가지고 있다
    
    캐시메모리를 사용하면 주 기억장치를 접근하는 횟수가 줄어들어 컴퓨터의 처리속도가 향상
    
    캐시 주소표는 검색시간을 단축시키기 위해 주로 연관 기억장치를 사용

순차코드(Sequence code) : 자료의 발생 순서, 크기 순서 등 일정 기준에 따라서 최초의 자료부터 차례로 일련 번호를 부여


표의숫자코드(significant digit code) : 코드화 대상 항목의 성질, 즉 길이, 넓이, 부피, 지름, 높이 등의 물리적 수치를 그대로 코드에 적용


블록코드(Block code) : 코드화 대상 항목 중에서 공통성이 있는 것끼리 블록으로 구분, 각 블록 내에서 일련 번호를 부여한느 방법


기호코드(mnemonic code) : 코드화 대상 항목의 명칭이나 약호와 관계있는 숫자나 문자, 기호를 이용하여 코드를 부여하는 방법(연상코드)


전위오류(Transposition Error) : 좌우 자리 바뀜  
필사 오류(Transcription Error) : 한자리 잘못  
임의 오류(Random Error) : 2가지 이상 오류 발생  
이중 오류(Double Transposition Error) : 전위 오류 2개 이상  


프로세스의 표준 처리 패턴 중 어떤 파일에서 특정한 조건에 만족하는 정보를 추출해 내는 처리는??

- 프로세스 표준 패턴


        매체변환(Conversion) : 사람이 판독가능 매체와 컴퓨터 저장장치인 파일매체 사이의 상호 간 변화  
        병합(Merge) : 두 개 이상 입력 자료 파일을 하나의 파일로 통합  
        갱신(Update) : 마스터파일 내용을 트랜잭션 파일 내용으로 변경해 새 마스터 파일을 생성  
        분류(Sort) : 비순서적 자료파일을 순서적(오름차순,내림차순)으로 나열  
        조합(Collate) : 파일 내 자료와 대조파일 내 자료가 같은 것끼리 모아 새로운 파일을 만드는 것  
        추출(Extract) : 자료 파일 내 있는 자료 중 조건 맞는 것만 추출  
        분배(Distribution) : 조건 맞는것과 맞지 않는 것으로 분리  


데이터파일 종류  

- 원시자료파일(Source Data File) : 원시데이터를 적당한 매체에 수록한 입력 데이터 파일  
- 마스터파일(Master File) : 전표처리에서 원장 또는 대장에 해당하는 파일  - 트랜잭션파일(Transaction File) : 마스터파일을 갱신 또는 조회(Update)  
- 히스토리파일(History File) : 사고 발생 시 , 데이터를 원상 복구 위한 보관용 파일  
- 요약파일(History File) : 편리한 집계 위해 다른 파일의 합계를 모아둔 파일  
- 트레일러 파일(Trailer File) : 마스터파일을 양방향레코드로 작성하여 2개로 작성한 파일  


부호 검사(Sign Check = Plus-Minus Check) : 계산 결과가 양수 또는 음수인지를 검사  
중복 레코드 검사(Double Record Check) : 계산을 처리하는 과정에서 동일한 레코드가 있는지를 검사  
불일치 레코드 검사(Unmatch Record Check) : 마스터 파일과 트랜잭션 파일을 조합할 때 키 항목이 일치하는지의 여부를 검사  
오버플로 검사(Overflow Check) : 계산 결과가 규정된 자릿수 또는 한계를 초과하는지를 검사  
제로 균형 검사(Zero Balance Check) : 계산 결과가 0이 되는지를 검사  
불능 검사(impossible Check) : 0으로 나누는 경우가 있는지를 검사
